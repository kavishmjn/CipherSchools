# AMCAT DATA-SET

The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). The data set contains data mainly about students with engineering disciplines. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset contains around 40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate. 

The dependent variables are : Salary, JobCity  and  Designation

Demographic Features: DOB,Gender
Academic Features: 10percentage,10board,12percenatge,12board,12graduation
College Features: CollegeCityTier,CollegeTier,CollegeGPA,CollegeID,CollegeCityID,Degree,Specialization,CollegeState
Independent Features:
Cognitive: Logical,English,Quant
Technical:Domain,ComputerProgramming',ElectronicsAndSemicon,ComputerScience,MechanicalEngg,ElectricalEngg,TelecomEngg,CivilEngg
Psychological:conscientiousness,agreeableness,extraversion,nueroticism,openess_to_experience

# BIVARIATE ANALYSIS

### "Relationship between non-psychological parameters and the salary"

```python
df1=df.loc[:,['10percentage','12percentage','Quant','English','Logical','Gender','Domain','Salary']]
df1['Total'] = df.loc[:,'English':'CivilEngg'].sum(axis=1)
sns.pairplot(df1,hue="Gender",palette='husl')
```

All the points are collected together. Hence the scatter plots are not helpful. Therefore we will use the heatmap to understand the relationship between the variables.

```python
sns.heatmap(df1.corr(),annot=True)
```

Salary has the best relationship with marks in Quant section while the worst relationship with the domain marks.
Overall the best relationship is between 10percentage and 12 percentage.

```python
sns.kdeplot(df1.Quant,df.Salary,cbar=True,shade=True)
plt.ylim(50000,700000)
plt.show()
```

Most of the candidates have got a salary between  3 - 4 lacs and their marks in quant. test is between 500-600.

```python
sns.kdeplot(df1.Total,df.Salary,cbar=True,shade=True)
plt.ylim(50000,700000)
plt.show()
```

Most of the candidates have got a salary between 3 - 4 lacs and their total marks in test  between 2000-2500.

### "Relationship between non-psychological parameters and the salary"

```python
df2=df.loc[:,'conscientiousness':'openess_to_experience']
df2['PsychTotal'] = df2.sum(axis=1)
df2['Salary']=df['Salary']
df2['Gender']=df['Gender']
sns.pairplot(df2,hue="Gender")
```

All the points are collected together. Hence the scatter plots are not helpful. Therefore we will use the heatmap to understand the relationship between the variables.

```python
sns.heatmap(df2.corr(),annot=True)
```

Salary has the best relationship with marks in agreeableness section while the worst relationship with the conscientiousness marks. Overall the best relationship is between Total Marks in PSychological test and 12 openess to expirience.

```python
sns.kdeplot(df2.agreeableness,df2.Salary,cmap="Reds",cbar=True,shade=True)
plt.xlim(-4,3)
plt.ylim(100000,700000)
plt.show()
```

Most of the candidates have got a salary between 3 - 4 lacs and their total marks in agreeableness test between -0.5 to 1

```python
sns.kdeplot(df2.PsychTotal,df2.Salary,cmap="Reds",cbar=True,shade=True)
plt.xlim(-7,7)
plt.ylim(100000,700000)
plt.show()
```

Most of the candidates have got a salary between 3 - 4 lacs and their total marks in non-psych test between -0.1 to 0.5.
